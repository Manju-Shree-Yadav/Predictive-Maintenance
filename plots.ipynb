```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load real-time sensor data from CSV
real_time_sensor_data = pd.read_csv('real_time.csv')

# Convert timestamp column to datetime format
real_time_sensor_data['Timestamp'] = pd.to_datetime(real_time_sensor_data['Timestamp'])

# Set up subplots
fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=True)

# Plot Vibration data
axs[0].plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Vibration'], color='blue')
axs[0].set_ylabel('Vibration')
axs[0].set_title('Real-time Sensor Data')
axs[0].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))

# Plot Moisture data
axs[1].plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Moisture'], color='green')
axs[1].set_ylabel('Moisture')
axs[1].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))

# Plot Temperature data
axs[2].plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Temperature'], color='red')
axs[2].set_ylabel('Temperature')
axs[2].set_xlabel('Timestamp')
axs[2].xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M:%S'))

# Format x-axis for all subplots
plt.setp(axs[-1].xaxis.get_majorticklabels(), rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Path to the real-time sensor data CSV file
csv_file_path = 'real_time.csv'

def plot_real_time_data(csv_file_path):
    plt.figure(figsize=(10, 6))
    plt.ion()  # Turn on interactive mode
    
    while True:
        clear_output(wait=True)
        
        # Read the latest data from the CSV file
        real_time_sensor_data = pd.read_csv(csv_file_path)
        
        # Plot the data
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Temperature'], label='Temperature')
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Vibration'], label='Vibration')
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Moisture'], label='Moisture')
        plt.xlabel('Timestamp')
        plt.ylabel('Value')
        plt.title('Real-time Sensor Data')
        plt.legend()
        plt.grid(True)
        plt.show()
        
        # Add a delay before the next plot
        time.sleep(1)  # Adjust the delay as needed

# Call the function to plot real-time data
plot_real_time_data(csv_file_path)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# Path to the real-time sensor data CSV file
csv_file_path = 'real_time.csv'

def plot_real_time_data(csv_file_path):
    plt.figure(figsize=(18, 12))
    
    while True:
        clear_output(wait=True)
        
        # Read the latest data from the CSV file
        real_time_sensor_data = pd.read_csv(csv_file_path)
        
        # Plot Temperature
        plt.subplot(3, 1, 1)
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Temperature'], color='red')
        plt.xlabel('Timestamp')
        plt.ylabel('Temperature')
        plt.title('Real-time Temperature')
        plt.grid(True)
        
        # Plot Vibration
        plt.subplot(3, 1, 2)
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Vibration'], color='blue')
        plt.xlabel('Timestamp')
        plt.ylabel('Vibration')
        plt.title('Real-time Vibration')
        plt.grid(True)
        
        # Plot Moisture
        plt.subplot(3, 1, 3)
        plt.plot(real_time_sensor_data['Timestamp'], real_time_sensor_data['Moisture'], color='green')
        plt.xlabel('Timestamp')
        plt.ylabel('Moisture')
        plt.title('Real-time Moisture')
        plt.grid(True)
        
        # Display the plots
        plt.tight_layout()
        plt.show()
        
        # Add a delay before the next plot
        time.sleep(1)  # Adjust the delay as needed

# Call the function to plot real-time data
plot_real_time_data(csv_file_path)
```
