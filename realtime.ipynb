```python
import numpy as np
import pandas as pd
import datetime
import time

# Function to generate synthetic sensor data
def generate_sensor_data(temperature_threshold=80, vibration_threshold=0.6):
    # Generate timestamps for the data
    timestamp = datetime.datetime.now()

    # Generate synthetic sensor data
    vibration_data = np.random.normal(0.5, 0.1)
    moisture_data = np.random.normal(40, 5)
    temperature_data = np.random.normal(70, 5)

    # Add failure labels and error types based on conditions
    failure_label = 0 if temperature_data <= temperature_threshold and vibration_data <= vibration_threshold else 1
    error_type = 'None' if failure_label == 0 else ('High Temperature' if temperature_data > temperature_threshold else 'High Vibration')

    # Create a DataFrame to store the sensor data with failure labels and error types
    sensor_data = pd.DataFrame({
        'Timestamp': [timestamp],
        'Vibration': [vibration_data],
        'Moisture': [moisture_data],
        'Temperature': [temperature_data],
        'FailureLabel': [failure_label],
        'ErrorType': [error_type]
    })

    return sensor_data

# Define failure thresholds
temperature_threshold = 80
vibration_threshold = 0.6

# Continuous generation of sensor data
while True:
    # Generate real-time sensor data
    real_time_sensor_data = generate_sensor_data(temperature_threshold, vibration_threshold)

    # Append the data to an existing CSV file or perform other operations as needed
    with open('real_time.csv', 'a') as f:
        real_time_sensor_data.to_csv(f, header=f.tell()==0, index=False)  # Append without header if file already exists

    # Display the generated sensor data
    print(real_time_sensor_data)

    # Wait for a specified interval (e.g., 1 second) before generating the next data
    time.sleep(1)
```
